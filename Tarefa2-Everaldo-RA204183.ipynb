```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### Exibindo o DATASET

```python
dataset = pd.read_csv('./datasets/medications.csv',on_bad_lines='skip')
print (type(dataset))
dataset
```

### 1) Histograma

    Foi realizado um filtro no conjunto de dados a fim de avaliar o custo total com o tratamento dentre os pacientes acometidos por COVID-19.

```python
dsFiltered = dataset.query("REASONDESCRIPTION == 'COVID-19'")
dsFiltered['TOTALCOST'].plot.hist(title='Custo Total em $')
```

### 2) Gráfico de Pizza

   Foi utilizado o conjunto de dados filtrado no item 1 para avaliar apenas os registros relacionados ao emprego de medicamentos cuja prescrição objetivava o tratamento da COVID-19.
 O gráfico exibe o percentual dos **10** medicamentos mais utilizados no tratamento de pacientes com COVID-19.

```python
dsFiltered.groupby(['DESCRIPTION']).DESCRIPTION.count().sort_values()[-10:].plot(kind='pie',autopct='%0.2f%%',label='',title='Medicamentos')
```

### 3) Gráfico de Dispersão (scatter plot)
    O objetivo é relacionar os custos totais com o motivo da prescrição (REASONDESCRIPTION). Assim, torna-se possível analisar os custos do tratamento medicamentoso de uma determinada doença em relação às outras.


```python
dataset = dataset.dropna() 
df2 = dataset.groupby('REASONDESCRIPTION')['TOTALCOST'].sum().sort_values()[:20].reset_index()
plt.xticks(rotation=90)
plt.scatter(df2.REASONDESCRIPTION, df2['TOTALCOST'])
```
